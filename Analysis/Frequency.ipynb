```python
file = open('A2-w-N2_ALL.txt', 'r') # Read .hb2 files
data = file.readlines()
```

```python
indx = []
for d in data:
    #print (d.strip())
    if 'pdb' in d:
        i = data.index(d)
        indx.append(i)
indx.append(len(data))

pdbwise = []
for i in range(len(indx)-1): 
    x = list(data[indx[i]:indx[i+1]])
    pdbwise.append(x)
        
```

```python
for p in pdbwise:
    count_all = 0
    count_nil = 0
    for n in p:
        if 'pdb' in n: 
            p_name = ((n.split('pdb')[1].split('.hb2')[0]).upper())
        if 'HOH' in n:
            count_all += 1
        if 'HOH' in n and 'NIL' in n:
            count_nil += 1
    if count_nil != 0:
        print(p_name + ' (' + str(count_nil) + ')')
            
        
        

        
```
